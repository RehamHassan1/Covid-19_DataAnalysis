```python
df['date'] = pd.to_datetime(df['date'])
global_trends = df.groupby('date').sum()
plt.figure(figsize=(12, 6))
plt.plot(global_trends.index, global_trends['new_cases'], label='New Cases', color='blue')
plt.plot(global_trends.index, global_trends['new_deaths'], label='New Deaths', color='red')
plt.title('Global New Cases and New Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

Total COVID-19 Cases & Deaths by Continent


```python
continent_summary = df.groupby('continent').agg({
    'total_cases': 'sum',
    'total_deaths': 'sum',
    'total_vaccinations': 'sum',
    'population': 'sum'
}).reset_index().sort_values(by='total_cases')
plt.figure(figsize=(7, 6))
sns.barplot(x='continent', y='total_cases', data=continent_summary.sort_values(by='total_cases'), palette='Blues_d')
plt.title('Total COVID-19 Cases by Continent')
plt.show()

plt.figure(figsize=(7, 6))
sns.barplot(x='continent', y='total_deaths', data=continent_summary.sort_values(by='total_deaths'), palette='Blues_d')
plt.title('Total COVID-19 Deaths by Continent')
plt.show()


```

Which continent had the highest number of total COVID-19 cases and deaths?

```python
continent_summary = df.groupby('continent').agg({
    'total_cases': 'sum',
    'total_deaths': 'sum'
}).reset_index().sort_values(by='total_cases')

print(continent_summary)
```

How did the number of new COVID-19 cases change globally over time?

```python
global_trends = df.groupby('date')['new_cases'].sum()

plt.figure(figsize=(10, 6))
plt.plot(global_trends.index, global_trends.values, color='blue')
plt.title('Global Trend of New COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.show()
```

Which countries have the highest number of COVID-19 tests performed per capita?

```python
top_testing_countries = df[['location', 'total_tests_per_thousand']].dropna().sort_values(by='total_tests_per_thousand', ascending=False)

print(top_testing_countries.head(10))
```

How did vaccination progress over time globally?

```python
vaccination_trends = df.groupby('date')['total_vaccinations'].sum()

plt.figure(figsize=(10, 6))
plt.plot(vaccination_trends.index, vaccination_trends.values, color='green')
plt.title('Global Progress of COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.show()
```

Life Expectancy vs COVID-19 Deaths per Million:




```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='life_expectancy', y='total_deaths_per_million', hue='continent', data=df, palette='tab10')
plt.title('Life Expectancy vs COVID-19 Deaths per Million')
plt.xlabel('Life Expectancy')
plt.ylabel('Total Deaths per Million')
plt.show()
```

Population Density vs Total Cases per Million

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='population_density', y='total_cases_per_million', data=df, hue='continent', palette='tab10')
plt.title('Population Density vs Total Cases per Million', fontsize=16)
plt.xlabel('Population Density', fontsize=12)
plt.ylabel('Total Cases per Million', fontsize=12)
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

GDP per Capita vs Total Deaths per Million

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_per_capita', y='total_deaths_per_million', data=df, hue='continent', palette='rocket')
plt.title('GDP per Capita vs COVID-19 Deaths per Million', fontsize=16)
plt.xlabel('GDP per Capita', fontsize=12)
plt.ylabel('Total Deaths per Million', fontsize=12)
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

Correlation Heatmap

```python
cols_of_interest = ['gdp_per_capita', 'population_density', 'total_cases_per_million', 'total_deaths_per_million']

plt.figure(figsize=(10, 8))
sns.heatmap(df[cols_of_interest].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

Total COVID-19 Deaths

```python
import plotly.express as px
total_deaths = df.groupby('location')['total_deaths'].max().reset_index()

fig = px.choropleth(total_deaths,
                    locations='location',
                    locationmode='country names',
                    color='total_deaths',
                    color_continuous_scale='matter',
                    labels={'total_deaths': 'Total Deaths'},
                    title='Total COVID-19 Deaths by Country')

fig.show()
```
